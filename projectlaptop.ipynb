```python
avg_price = df.groupby('brand')['price'].mean().sort_values()
avg_price.plot(kind='bar')
plt.title("sorted Brand Distribution")
plt.xlabel("Brand")
plt.ylabel("Price")
plt.show()
```

```python
sns.countplot(x = "warranty",hue = "Ram",data = df)
plt.title("Warranty according to Ram")
plt.show()
```

```python
plt.Figure(figsize=(10,5))
sns.boxplot(x = 'brand',y = 'price',data = df)
plt.xticks(rotation = 90)
plt.title("Brand vs price")
plt.show()
```

```python
count = df['warranty'].value_counts()
plt.pie(count,labels=count.index,autopct= '%1.1f%%')
plt.title("Percent of warranty")
plt.show()
```

```python
predict = df.groupby("warranty")[[
   
   "display_size",
   "resolution_width",
   "resolution_height",
   "spec_rating"
]].mean()

predict.plot(kind="bar")
plt.title(" Laptop Estimation")
plt.ylabel("Average Prediction")
plt.xticks(rotation = 0)
plt.show()
```
